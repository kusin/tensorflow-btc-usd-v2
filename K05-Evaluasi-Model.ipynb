```python
# lib manipulation dataset
import pandas as pd
import numpy as np
```

```python
# lib data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def barplot(x, y, title):
  
  # create figure
  fig, ax = plt.subplots(figsize = (8,4))
  plt.bar(x, y, width=0.75, color=['#0080bf', '#ffb703'])

  # set labels
  ax.set_title(title,fontsize=12)
  ax.set_xlabel("",fontsize=12)
  ax.set_ylabel("",fontsize=12)
  ax.grid(True)

  # show plot
  plt.show()
```

```python
# Hasil R
df = pd.DataFrame({
  "model": ['SBi-LSTM-RNN', 'SBi-GRU-RNN'],
  "akurasi": [0.9962, 0.9975]
})

# Hasil R
barplot(df["model"], df["akurasi"], "Evaluation models with correlation R")
```

```python
# Hasil MAE
df = pd.DataFrame({
  "model": ['SBi-LSTM-RNN', 'SBi-GRU-RNN'],
  "akurasi": [1149.0806, 1077.2294]
})

# Hasil MAE
barplot(df["model"], df["akurasi"], "Evaluation models with correlation MAE")
```

```python
# Hasil RMSE
df = pd.DataFrame({
  "model": ['SBi-LSTM-RNN', 'SBi-GRU-RNN'],
  "akurasi": [1696.8877, 1575.4796]
})

# Hasil RMSE
barplot(df["model"], df["akurasi"], "Evaluation models with correlation RMSE")
```

```python
# Hasil MAPE
df = pd.DataFrame({
  "model": ['SBi-LSTM-RNN', 'SBi-GRU-RNN'],
  "akurasi": [0.0351, 0.0299]
})

# Hasil MAPE
barplot(df["model"], df["akurasi"], "Evaluation models with correlation MAPE")
```
