```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def barplot(data, title, ymin, ymax, step):

   # Define the x-axis labels and the width of each bar
   labels = ['8', '16', '24', '32',]
   bar_width = 0.25

   # Create a numpy array of the x-axis positions for each group of bars
   x_pos = np.arange(len(labels))

   # Create a figure and axis object
   fig, ax = plt.subplots(figsize = (8,4))

   # Loop through each group of bars and create a set of bars for each group
   for i, (group, values) in enumerate(data.items()):
      # Calculate the x-axis position for the current group of bars
      pos = x_pos + (i * bar_width)
      # Create a set of bars for the current group
      ax.bar(pos, values, width=bar_width, label=group)

   # Set the x-axis labels and tick positions
   ax.set_xticks(x_pos + ((len(data) - 1) / 2) * bar_width)
   ax.set_xticklabels(labels)

   # Add a legend to the plot
   ax.legend(loc='best', ncols=3)
   ax.grid(True)

   # Add axis labels and a title to the plot
   ax.set_title(title, fontsize=12)
   ax.set_xlabel('Batch Size', fontsize=10)
   ax.set_ylabel('MAPE', fontsize=10)

   # Membulatkan batas y ke 2 angka di belakang koma
   # ymin, ymax = plt.ylim()
   # ymin_rounded = round(ymin, 2)
   # ymax_rounded = round(ymax, 2)
   #ax.set_ylim([ymin, ymax])
   ax.set_yticks(np.arange(ymin, ymax, step))

   # Show the plot
   #plt.tight_layout()
   plt.show()
```

```python
def lineplot(batch_sizes, adam, adamax, rmsprop, title):
  # lineplot
  fig, ax = plt.subplots(figsize = (8,4))
  ax.plot(batch_sizes, adam, marker='o', color='tab:blue', label='Adam')
  ax.plot(batch_sizes, adamax, marker='o', color='tab:orange', label='Adamax')
  ax.plot(batch_sizes, rmsprop, marker='o', color='tab:green', label='RMSprop')

  # add legend
  ax.legend(loc='best', ncols=3)
  ax.grid(True)

  # add labels
  ax.set_title(title, fontsize=12)
  ax.set_xlabel('Batch Size', fontsize=10)
  ax.set_ylabel('Time (seconds)', fontsize=10)

  # custom labels
  ax.set_xticks(np.arange(8, 36, 8))
  ax.set_yticks(np.arange(0, 550, 100))

  # show plot
  #plt.tight_layout()
  plt.show()
```

```python
# with epoch 25
batch_sizes = [8, 16, 24, 32]
adam    = [218.33, 106.54, 63.94, 52.32]
adamax  = [218.26, 115.05, 62.66, 52.03]
rmsprop = [220.78, 115.70, 63.47, 53.63]
title   = ""

# SBi-LSTM with epoch 25
lineplot(batch_sizes, adam, adamax, rmsprop, "Computation time in SBi-LSTM with epoch 25")
```
