```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(time_values, ylabel):
  # Data dari tabel
  epochs = [25, 50]
  batch_sizes = [8, 16, 24, 32]
  optimizers = ['Adam', 'Adamax', 'RMSprop']

  # Membuat plot
  fig, axes = plt.subplots(1, 2, figsize=(14,4), sharey=True)

  for i, epoch in enumerate(epochs):
    ax = axes[i]
    for j, optimizer in enumerate(optimizers):
      times = [time_values[epoch][bs][j] for bs in batch_sizes]
      ax.plot(batch_sizes, times, marker='o', label=optimizer)
    ax.set_title(f'Epoch {epoch}')
    ax.set_xlabel('Batch Size')
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(True)

  plt.tight_layout()
  plt.show()
```
