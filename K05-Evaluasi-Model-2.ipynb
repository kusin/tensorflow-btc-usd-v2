```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(time_values, ylabel):
  # Data dari tabel
  epochs = [25, 50]
  batch_sizes = [8, 16, 24, 32]
  optimizers = ['Adam', 'Adamax', 'RMSprop']

  # Membuat plot
  fig, axes = plt.subplots(1, 2, figsize=(14,4), sharey=True)

  for i, epoch in enumerate(epochs):
    ax = axes[i]
    for j, optimizer in enumerate(optimizers):
      times = [time_values[epoch][bs][j] for bs in batch_sizes]
      ax.plot(batch_sizes, times, marker='o', label=optimizer)
    ax.set_title(f'Epoch {epoch}')
    ax.set_xlabel('Batch Size')
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(True)

  plt.tight_layout()
  plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
batch_sizes = [8, 16, 24, 32]
adam_25 = [0.0612, 0.0867, 0.0876, 0.0718]
adamax_25 = [0.0778, 0.0310, 0.0526, 0.1003]
rmsprop_25 = [0.0741, 0.0434, 0.0398, 0.0485]

adam_50 = [0.1130, 0.0442, 0.0221, 0.0289]
adamax_50 = [0.0806, 0.0264, 0.0403, 0.0718]
rmsprop_50 = [0.0543, 0.0364, 0.0345, 0.0336]

bar_width = 0.25
index = np.arange(len(batch_sizes))

def plot_grouped_bar(data, title):
    plt.figure(figsize=(8,4))
    plt.bar(index, data[0], bar_width, label='Adam')
    plt.bar(index + bar_width, data[1], bar_width, label='Adamax')
    plt.bar(index + 2 * bar_width, data[2], bar_width, label='RMSprop')

    # set labels
    plt.title(title)
    plt.xticks(index + bar_width, batch_sizes)
    plt.xlabel("Batch Size")
    plt.ylabel("MAPE")
    plt.ylim(0,0.15)
    plt.legend()
    plt.grid(True)

    # show plot
    plt.tight_layout()
    plt.show()

# Plot for Epoch 25
plot_grouped_bar([adam_25, adamax_25, rmsprop_25], 'SBi-GRU Algorithm with epoch 25')
# Plot for Epoch 50
plot_grouped_bar([adam_50, adamax_50, rmsprop_50], 'SBi-GRU Algorithm with epoch 50')
```

```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(8,4))

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the plot
data = {
   'Adam': [20, 35, 30, 35,],
   'Adamax': [25, 32, 34, 20,],
   'RMSprop': [12, 20, 22, 30,],
}

# Define the x-axis labels and the width of each bar
labels = ['8', '16', '24', '32',]
bar_width = 0.2

# Create a numpy array of the x-axis positions for each group of bars
x_pos = np.arange(len(labels))

# Create a figure and axis object
fig, ax = plt.subplots(figsize = (8,4))

# Loop through each group of bars and create a set of bars for each group
for i, (group, values) in enumerate(data.items()):
   # Calculate the x-axis position for the current group of bars
   pos = x_pos + (i * bar_width)
   # Create a set of bars for the current group
   ax.bar(pos, values, width=bar_width, label=group)

# Set the x-axis labels and tick positions
ax.set_xticks(x_pos + ((len(data) - 1) / 2) * bar_width)
ax.set_xticklabels(labels)

# Add a legend to the plot
ax.legend()

# Add axis labels and a title to the plot
ax.set_xlabel('Batch Size')
ax.set_ylabel('MAPE')
ax.set_title('Grouped Bar Plot')
ax.legend(loc='upper left', ncols=3)

# Show the plot
plt.show()
```
